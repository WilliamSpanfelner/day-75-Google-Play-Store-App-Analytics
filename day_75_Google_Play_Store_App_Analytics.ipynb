# Find the Most Expensive Apps, Filter out the Junk, and Calculate a (ballpark) Sales Revenue Estimate

Let's examine the Price column more closely.

**Challenge**: Convert the price column to numeric data. Then investigate the top 20 most expensive apps in the dataset.

Remove all apps that cost more than $250 from the `df_apps_clean` DataFrame.

Add a column called 'Revenue_Estimate' to the DataFrame. This column should hold the price of the app times the number of installs. What are the top 10 highest grossing paid apps according to this estimate? Out of the top 10 highest grossing paid apps, how many are games?


```python
df_apps_clean.head()
```

> Converting the Price column from str type to numeric requires that the dollar sign is removed, firstly.  My approach to this was list comprehension.  A price_col variable is created by iterating over the column and removing the $ from the data.

> Then the price column data is converted into float type.

```python
price_col = [item.replace('$', '') for item in df_apps_clean.Price]
print(type(price_col[0]))
price_col = [float(item) for item in price_col]
type(price_col[0])
```

> Finally, the price column gets inserted into the dataframe.

```python
df_apps_clean.insert(1, 'Sub', price_col)
df_apps_clean.sort_values('Sub', ascending=False).head(20)
```

> Alternatively, the Price column can be converted inplace.

```python
df_apps_clean.Price = df_apps_clean.Price.astype(str).str.replace('$', '')
df_apps_clean.Price = pd.to_numeric(df_apps_clean.Price)
df_apps_clean.sort_values('Price', ascending=False).head(20)
```

### The most expensive apps sub $250

> Taking an example from the documentation, the following series can be 'filtered' by a specified value using the lt function. 

```python
a = pd.Series([1000, 1000, 100, 249, 111], index=['a', 'b', 'c', 'd', 'e'])
a
```

> To view only the values less than 250, use that as the key to the list values.

```python
a[a.lt(250)]
```

> Applying the same technique to the issue at hand...

```python
df_apps_clean[df_apps_clean.Sub.lt(250)].sort_values(['Sub', 'Installs (int)'], ascending=False).head(20)
```

> A slightly different approach is to redefine the df_apps_clean dataframe as below.

```python
df_apps_clean = df_apps_clean[df_apps_clean['Price'] < 250]
df_apps_clean.sort_values('Price', ascending=False).head(5)
```

### Highest Grossing Paid Apps (ballpark estimate)

> My approach to finding the highest grossing apps was to define a new variable, multiply the installs by the amount paid and inserting that column into the dataframe.
> Finally, the top 10 rows were skimmed off.

```python
revenue_est_col = df_apps_clean['Sub'] * df_apps_clean['Installs (int)']
revenue_est_col.sort_values(ascending=False).head()
df_apps_clean.insert(1, 'Revenue_Est', revenue_est_col)
```

```python
top_10_grossing = df_apps_clean[df_apps_clean.Sub.lt(250)].sort_values('Revenue_Est', ascending=False).head(10)
top_10_grossing
```

> To find the number of the top 10 that are games can be found by the len method of the resulting list.

```python
len(top_10_grossing[top_10_grossing.Category.eq('GAME')])
```

> There is always another way...

```python
df_apps_clean['Revenue_Estimate'] = df_apps_clean['Installs (int)'].mul(df_apps_clean.Price)
df_apps_clean.sort_values('Revenue_Estimate', ascending=False)[:10]
```

# Extracting Nested Data from a Column

**Challenge**: How many different types of genres are there? Can an app belong to more than one genre? Check what happens when you use .value_counts() on a column with nested values? See if you can work around this problem by using the .split() function and the DataFrame's [.stack() method](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.stack.html). 


> Find how many different genre types there are by first splitting apart any compound genres and subsequently adding them to a set.

```python
x = [1, 2, 20, 6, 210]
x += x
print(x)
set(x)
```

```python
genres = ['Medical', 'Arcade', 'Arcade', 'Arcade', 'Arcade', 'Medical', 'Productivity', 'Video Players & Editors', 'Editors & Video Players']
set_genres = set()
for genre in genres:
    if not ' & ' in genre:
        set_genres.add(genre)
    set_genres.update(genre.split(sep=' & '))
print(len(set_genres))
set_genres
```

> The value_counts function is treating strings with a semicolon as a unique item.  The semicolon needs to be separated from the string in which it is present.

```python
genre_count = df_apps_clean.Genres.value_counts().sort_values(ascending=True)[:10]
genre_count
```

> The original Genres column has 8184 rows of data.

```python
df_apps_clean.Genres.shape
```

> Splitting the strings apart with the split method using the ";" as a separator and expanding the resulting strings into separate columns, will make compound genres into individual ones.  Finally, the stack method will make the columns into a list.
Now the unique values of genres may be counted.

```python
genres = df_apps_clean.Genres.str.split(';', expand=True).stack()
print(f'The genres column has the following shape: {genres.shape}')

count_genres = genres.value_counts()
print(f'Number of genres: {len(count_genres)}')
```
